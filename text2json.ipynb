```python
import json
import glob
import re
```

```python
files = glob.glob("tesseract-ocr/*.txt")
len(files)
```

```python
for filename in files:
    all_data = {"header": "",
               "segments": []}
    with open (filename, "r", encoding="utf-8") as f:
        data = f.read()
    while "\n \n" in text:
        text = text.replace("\n \n", "\n")
    while "\n\n" in text:
        text = text.replace("\n\n", "\n")
    while " \n" in text:
        text = text.replace(" \n", "\n")
    text = text.replace("-\n", "\n")
    
    #removes header and page number
    cleaned = re.sub(r"USHMM Archives RG-([^\s]+) \d{1,2}", "", data)
    
    #removes timestamps
    cleaned = re.sub(r"\d{1,2}:\d{1,2}:\d{1,2}", "", data)
    while "\n\n" in cleaned:
        cleaned = cleaned.replace("\n\n", "\n")

    new = re.split('Question:|Q:', cleaned)
    all_data["header"] = new[0]
    questions = new[1:]
    x=1
    for question in questions:
        
        #removes some other artifacts of the pdf
        question = question.replace("\n-", "").replace("\n", " ")
        question = question.replace("http://collections.ushmm.org Contact reference@ushmm.org for further information about this collection", "")
        question = question.replace("This is a verbatim transcript of spoken word. It is not the primary source, and it has not been checked for spelling or accuracy.", "")
   
        while "  " in question:
            question = question.replace("  ", " ")
        result = re.split('Answer:|A:', question)
        if len(result) == 2:
            q, a = result
            all_data["segments"].append({"num": x, 
                                         "question": q.strip(),
                                         "answer": a.strip()}
                                       )
        else:
            all_data["segments"].append({"num": x,
                                         "validation_needed": True,
                                         "data": result}
                                       )
            print (filename)
        x=x+1
    json_file = filename.replace(".txt", ".json").replace("tesseract-ocr", "json")
    with open (json_file, "w") as f:
        json.dump(all_data, f, indent=4)
            
    
```
